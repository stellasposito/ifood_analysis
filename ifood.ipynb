## Análise exploratória Ifood
Esse dataset consta com informações sobre clientes do Ifood, o qual foi fornecido pelo curso EBA - Estatística do Básico ao Avançado.



Colunas que foram realizadas análises:

**Income:** salário do usuário

**education_level:** nível de educação do usuário

**marital_status:** estado civil

**kids:** quantidades de filhos (0 a 3)

**expenses:** valor das despesas dos usuários

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mkt_data.csv')
```

```python
data
```

```python
# verificando as informações das colunas
data.info()
```

```python
#verificando as colunas com dados numéricos
num = data.select_dtypes(include=['float64', 'int64']).columns
print('Colunas com dados numéricos:')

for coluna in num:
    print(coluna)
```

```python
#itens duplicados
duplicated = data.duplicated().sum()
print(duplicated)
```

```python
#dados nulos
nulos = data.isnull().sum()
print(nulos)
```

```python
data.head(10)
```

```python
#descrição estatística de cada coluna
data.describe()
```

```python
#excluir coluna Unnamed: 0 e colunas com dados nulos
df = data.drop(columns=['Unnamed: 0'])
df = df.dropna(axis=1, how='any')

```

```python
columns = df.columns

for i in columns:
    print(i)
```

```python
df
```

## Análises com a coluna Income

```python
df['Income'].max()
```

```python
df['Income'].min()
```

```python
sns.boxplot(df.Income)
```

```python
#transformando a variável em número inteiro
df['Income']=df['Income'].astype(int)
```

```python
df['Income'].dtype
```

```python
#dividindo a variável por 12 para facilitar a visualização do salário
df['Income'] = df['Income'] / 12
```

```python
sns.boxplot(df.Income)
```

```python
df['Income'].describe()
```

```python
# criando uma coluna que define quem ganha bem, e quem não ganha bem, tendo como base o 3º quartil: 5690.083333
df['Categoria_Salario'] = np.where(df['Income']>5690.083333, "ganham_bem", "nao_ganham_bem")
```

```python
df.head(20)
```

```python
count = df[''].value_counts()
print(count)
```

```python
count = df['Categoria_Salario'].value_counts()

sns.barplot(x=count.index, y=count.values)
plt.xlabel('Categoria do Salário')
plt.ylabel('Frequência')
plt.title('Salários')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.show()
```

## Análises com a coluna Education Level

```python
contagem = df['education_level'].value_counts()
print(contagem)
```

```python
contagem = df['education_level'].value_counts()

sns.barplot(x=contagem.index, y=contagem.values)
plt.xlabel('Nível de Educação')
plt.ylabel('Frequência')
plt.title('Frequência dos Níveis de Educação')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.show()
```

```python
plt.hist(df['education_level'], bins =9)
```

```python
#agrupando as colunas education_level e Categoria_Salario para realizar análises gráficas
grouped = df.groupby(['education_level', 'Categoria_Salario']).size().unstack()
```

```python
#gráfico de barras agrupadas
grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Relação entre Nível de Educação e Categoria de Salário')
plt.xlabel('Nível de Educação')
plt.ylabel('Número de Usuários')

plt.show()
```

```python
#gráfico de barras paralelas para uma melhor visualização
plt.figure(figsize=(10, 6))
sns.countplot(x='education_level', hue='Categoria_Salario', data=df)

plt.title('Relação entre Nível de Educação e Categoria de Salário')
plt.xlabel('Nível de Educação')
plt.ylabel('Número de Usuários')
plt.xticks(rotation=45)
plt.legend(title='Categoria de Salário')

plt.show()
```

## Análises com a coluna Marital Status

```python
marital = df['marital_status'].value_counts()
print(marital)

sns.barplot(x=marital.index, y=marital.values)
plt.xlabel('Estado civil')
plt.ylabel('Frequência')
plt.title('Frequência de níveis de educação')
plt.xticks(rotation=45)
plt.show()

```

```python
count = df.groupby(['marital_status', 'kids']).size().unstack(fill_value=0)

# Plotando o gráfico de barras empilhadas
count.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Estado Civil')
plt.ylabel('Quantidade de pessoas')
plt.title('Distribuição do Número de Filhos por Estado Civil')
plt.xticks(rotation=45)
plt.legend(title='Número de Filhos')
plt.show()
```

```python
count.plot(kind='bar', stacked=False, figsize=(10, 6))

# Adicionando rótulos e título
plt.xlabel('Estado Civil')
plt.ylabel('Quantidade de pessoas')
plt.title('Distribuição do Número de Filhos por Estado Civil')
plt.xticks(rotation=45)
plt.legend(title='Número de Filhos')
plt.show()
```

## Análises com a coluna expenses

```python
df['kids']=df['kids'].astype(int)
```

```python
df['kids'].dtype
```

```python
#verificar se a quantidade de filhos interfere no quanto a pessoa gasta na plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(x='kids', y='expenses', data=df)

plt.title('Distribuição das Despesas por Número de Filhos')
plt.xlabel('Número de Filhos')
plt.ylabel('Despesas')

plt.show()
```

```python
#verificar se as pessoas que ganham mais gastam mais na plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(x='Categoria_Salario', y='expenses', data=df)

plt.title('Distribuição das Despesas por Categoria de Salário')
plt.xlabel('Categoria de Salário')
plt.ylabel('Despesas')

plt.show()
```
